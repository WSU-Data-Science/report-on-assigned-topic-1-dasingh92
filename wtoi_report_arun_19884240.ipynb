# Wave theory of information

***
***
Our sensing ability to capture the distinction between two different waveforms dictates the limits on the amount of information transferred by a propagating wave. The problem of quantifying this precisely requires a mathematical description and a physical understanding of both propagation and the communication processes <a id="ref-1" href="#cite-franceschetti_2017">(Massimo Franceschetti, 2017)</a>. The focus in this document will be on propagation of electromagnetic waves based on Maxwell's theory of electromagnetism, and on information theory as described by *Claude Shannon*.

Information theory as described by ***Claude Shannon*** in his famous paper <a id="ref-2" href="#cite-shannon_1948">(Shannon, 1948)</a> has always had a heavier background in mathematics and engineering. Every book and course in information theory is more often than not based on mathematical or engineeering viewpoint. 
The central theme of  <a id="ref-3" href="#cite-franceschetti_2017">(Massimo Franceschetti, 2017)</a> is that Shannon's information-theoretic limits are natural. They are revealed by observing physical quantities at certain asymptotic scales where finite dimenionality emerges and observational uncertainties are averaged out. These limits are rigorous, and obey the mathematical rules that govern the model of reality on which the the physical theories are based. The author attempts to fill the gaps left by lack of consideration of laws of physics in information theory especially for wave theory. This results in a beautiful weave between information theory and wave theory where the limits imposed by Shannon's information theory are shown as not just bounded by mathematical laws but also as a natural consequences of laws of physics. 

To get a clearer understanding of the topics involved in this document we will first introduce **Wave Theory** and **Information Theory** individually and then dive into the ***Wave theory of information***.

## Wave Theory

***

### Wave

In physics, mathematics, and other related fields *wave* is described as a propagating dynamic disturbance(change from equilibrium) of one or more quantities, sometimes described by a *wave equation*. In physical waves at least two field quantities in the wave medium are involved.

In an electromagnetic wave (the ones we are interested in from our viewpoint w.r.t wave theory of information), coupling between the electric and magnetic fields which sustains propagation of a wave involving these fields according to Maxwell's equations. Electromagnetic waves can travel through a vacuum and through some dielectric media (at wavelengths where they are considered transparent). Electromagnetic waves, according to their frequencies (or wavelengths) have more specific designations including radio waves, infrared radiation, terahertz waves, visible light, ultraviolet radiation, X-rays and gamma rays. 

Waves are of two types:
* **Transverse waves** : In these waves the disturbance in the medium is perpendicular to the motion of the wave. For example, ripples in water, Electromagentic waves, etc.

* **Longitudinal Waves**: In such waves the disturbance in the medium is in the same direction as the motion of the wave. For example, sound waves, Tsunamis, seismic waves, etc.

---

### Mathematical Description

#### Single Waves

A [wave](https://en.wikipedia.org/wiki/Wave) can be described as a function $F(x, t)$ where $x$ is *position* (usually a vector in the Cartesian three-dimensional space $\mathbb{R}^3$) and $t$ is *time*.

The value of $F ( x , t )$  can be any physical quantity of interest assigned to the point $x$  that may vary with time. For example, if $F$ represents the vibrations inside an elastic solid, the value of $F ( x , t )$ is usually a vector that gives the current displacement from $x$  of the material particles that would be at the point $x$ in the absence of vibration. For an electromagnetic wave, the value of $F$ can be the electric field vector $E$, or the magnetic field vector $H$, or any related quantity, such as the *Poynting vector* $E Ã— H$.

#### Wave Families

More often than not we need to understand a large set of possible waves. In these situations, one may describe such a family of waves by a function $F(A,B,...;x, t)$ that depends on certain parameters $A,B,...$ besides $x$ & $t$. Then one can obtain different waves -- that is, different functions of $x$ & $t$ - by choosing different values for those parameters.

#### Differential wave equations

Another way to describe and study a family of waves is to give a mathematical equation that, instead of explicitly giving the value of $F(x,t)$, only constrains how those values can change with time. Then the family of waves in question consist of all functions $F$ that satisfy those constraints - that is, all *solutions* of the equation.

For example, we can describe all possible sounds echoing within a container of gas by a function $F(x,t)$ that gives the pressure at a point $x$ and time $t$ within that container. If the gas was initially at uniform temperature and composition, the evolution of $F$ is constrained by the formula, 
\begin{equation}\label{eq:1}
\frac{\partial^2{F}}{\partial{t}^2}(x,t) = \alpha \Bigg(\frac{\partial^2{F}}{\partial{x^2_1}}(x,t) + \frac{\partial^2{F}}{\partial{x^2_2}}(x,t) + \frac{\partial^2{F}}{\partial{x^2_3}}(x,t) \Bigg) +\beta P(x,t)
\end{equation}

where $P(x,t)$ is some extra compression force that is being applied to the gas near $x$ by some external process, such as *loudspeaker* or *piston* right next to *p*.

This same equation differential equation describes the behavior of mechanical vibrations and electomagnetic fields in a homogeneous isotropic non-conducting solid. The above equation is also referred to as "the" wave equation in mathematics, although it describes a very specific family of waves.

### Waveform

In electronics, acoustics, and related fields, the waveform of a signal is the shape of its graph as a function of time, independent of its time and magnitude scales and of any displacement in time.
The following are the most common types of waveforms:
1. Sine Wave(t, $\lambda$, a, $\phi$) = $a sin\frac{2\pi - t}{\lambda}$

2. Square Wave $(t,\lambda ,a,\phi)$=\begin{cases}a,&(t-\phi ){\bmod {\lambda }}<{\text{duty}}\\-a,&{\text{otherwise}}\end{cases}.

This waveform is commonly used to represent digital information. A square wave of constant period contains odd harmonics that decrease at -6dB/octave.

3. Triangle Wave $(t, \lambda, a, \phi) = \frac{2a}{\pi} arcsin \ sin \frac{2 \pi t - \phi}{\lambda}$. It contains odd harmonics which decrease at -12 dB/octave. 

4. Sawtooth wave $(t, \lambda, a, \phi) = \frac{2a}{\pi} arctan \ tan \frac{2 \pi t - \phi}{2\lambda}$. This looks like the teeth of a saw. Found often in time bases for display scanning. It is used as the starting point for subtractive synthesis, as a sawtooth wave of constant period contains odd and even harmonics that decrease at âˆ’6 dB/octave.

```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
```

```python
freq = 2
amp = 1
time = np.linspace(0, 2, 2000)

# Generate sine wave
sin_wave = amp * np.sin((2 * np.pi * time * freq))

# Generate square wave
sq_wave = amp * sg.square(2*np.pi*freq*time, duty = 0.5)

# Generate triangle wave
tri_wave = amp*sg.sawtooth(2*np.pi*freq*time, width=0.5)

# Generate saw tooth wave
saw_wave = amp*sg.sawtooth(2*np.pi*freq*time, width = 0)

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (20,12))

ax[0][0].plot(time, sin_wave, color = "red")
ax[0][0].set_xlabel("Time")
ax[0][0].set_ylabel("Amplitude")
ax[0][0].set_title("Sine Wave")

ax[0][1].plot(time, sq_wave, color = "blue")
ax[0][1].set_xlabel("Time")
ax[0][1].set_ylabel("Amplitude")
ax[0][1].set_title("Square Wave")

ax[1][0].plot(time, tri_wave, color = "green")
ax[1][0].set_xlabel("Time")
ax[1][0].set_ylabel("Amplitude")
ax[1][0].set_title("Triangle Wave")

ax[1][1].plot(time, saw_wave, color = "purple")
ax[1][1].set_xlabel("Time")
ax[1][1].set_ylabel("Amplitude")
ax[1][1].set_title("Sawtooth Wave")
```

### Electromagnetic Radiation

Electromagnetic radiation consists of ***electromagnetic waves*** which are basically *synchronized* oscillations of **electric fields** and **magnetic fields**. In a homogeneous, isotropic media, the oscillations of the electric fields and magnetic fields are perpendicular to each other and also to the direction of propagation of light. .

Electromagnetic waves are emitted by charged particles undergoing acceleration and these waves can subsequently interact with other charged particles exerting force on them. An EM wave can carry energy, momentum, and angular momentum away from their source particle and can impart those quantities to matter with which they interact.<a id="ref-4" href="#cite-cloude_1996">(Cloude, 1996)</a>

#### Maxwell's Electromagentic wave Equations

**James Clerk Maxwell** derived a wave form of the electric and magnetic equations, thus uncovering the wave-like nature of electric and magnetic fields and their symmetry. Because the speed of EM waves predicted by the wave equation coincided with the measured speed of light, Maxwell concluded that light itself is an EM wave <a id="ref-5" href="#cite-maxwell_1865">Maxwell, 1865</a> (this is referred to as **Wave Theory** of light).  Maxwell's equations were confirmed by Heinrich Hertz through experiments with radio waves. 

The **electromagnetic wave equation** is a *second-order partial differential equation* that describes the propagation of electromagnetic waves through a medium or in a vacuum. It is a three-dimensional form of the wave equation. The homogeneous form of the equation, written in terms of either the electric field $E$ or the magnetic field $B$, takes the form,

\begin{equation}
\Bigg( v^2_{ph} \nabla^2 - \frac{\partial^2}{dt^2}\bigg) E = 0 
\end{equation}

\begin{equation}
 \Bigg( v^2_{ph} \nabla^2 - \frac{\partial^2}{dt^2}\bigg) B = 0 
\end{equation}


where

$ v_{ph} = \frac{1}{\sqrt{\mu\epsilon}}$

is the speed of light (i.e. phase velocity) in a medium with permeability $\mu$, and permittivity $\epsilon$, and $\nabla^2$ is the Laplace operator.

For derivation of the eq.(4) and eq.(5) using Maxwell's equations refer to Appendix A.

#### Near field and Far field EMR

Maxwell's equations establish the fact that some currents or charges (sources) produce an *EM field* near them that doesn't have the characteristics of EMR. For example, currents produce a magnetic field which is often of the *magnetic dipole* type and dies out as distance from source increases. These fields make up the **near-field** close to the EMR source and don't in anyway contribute to the EMR. They can efficiently tranfer power to receiver very near to the source such as *magnetic resonance* in a transformer or feedback behavior that happens to the coil of a *metal detector*. 

On the other hand the **EM far-field** consists of *radiation* that is independent of the transmitter in the sense that the transmitter requires the same power to send these changes out in the field, irrespective of the fact whether the signal is immediately picked up or not. This distant part of the electromagnetic field is "electromagnetic radiation" (also called the far-field). The far-fields propagate (radiate) without allowing the transmitter to affect them. This causes them to be independent in the sense that their existence and their energy, after they have left the transmitter, is completely independent of both transmitter and receiver. For example, radio waves, X-rays, gamma rays, etc.

Throughout this document when we say **EM waves** we mean **EM far-field radiations.**

## Information Theory

---

**Information theory** is the scientific study of *quantification, storage and communication of information*. The field was fundamentally established by the works of Harry Nyquist and Ralph Hartley, in the 1920s, and Claude Shannon in the 1940s.

A key measure in information theory is **entropy**. Entropy quantifies the amount of uncertainty involved in the value of a random variable or the outcome of a random process. For example, flipping a fair coin (having two equally likely outcomes) has low entropy(provides less information) as compared to roll of a dice(six equally likely outcomes).

The concrete establishment of information theory as a worldwide field of study was after the publication of Claude E. Shannon's classic paper "A Mathematical Theory of Communication" in the Bell System Technical Journal in July and October 1948 [(Shannon, 1948)](#cite-shannon_1948). This groundbreaking paper brough forward ideas such as:
* the information **entropy** and **redundancy** of a source , and it's relevance through the ***source coding theorem***,
* the mutual information and the channel capacity of a noisy channel, including the promise of a perfect loss-free communication given by the noisy-channel coding theorem,
* the practical results of **Shannon-Hartley law** for channel capacity of a guassian channel, and
* the bit - a new way of seeing the most fundamental unit of information.

### Quantities of information

Information theory is based on *probability theory* and statistics and it most concerns with measures of information of the distributions associated with random variables. Two most important quantities in information theory are:
1. **Entropy**: It is a measure of information in a single random variable. It is a property of the probability distribution of a random variable and provides us a limit on the rate at which data generated by independent variables of a given distribution can be reliably compressed.


2. **Mutual Information**: It is a measure of information in common between two random variables. It is a property of the joint distribution of the two random variables and gives use the maximum reliable communication across a noisy channel.

### Entropy of an information source

Based on [probability mass function](https://en.wikipedia.org/wiki/Probability_mass_function) of each source symbol to be communicated, the Shannon entropy $H$, in units of bits(per symbol), is given by

\begin{equation}
H = - \sum_i p_i \log_2(p_i)
\end{equation}

The above equation gives us entropy in terms of bits per symbol because of using $\log_2$. If we where to use $\log_n$ (natural log with base $e$, euler's number) we would get entropy in ***nats*** per symbol. This form is sometimes preferred as it simplifies the analysis by avoiding the need to include extra constants in the formulas. In $\log_{10}$ the measured entropy is in ***hartley*** per symbol. 


Intuitively, if one transmits say 256 bits of information and the value of these bits are known at the receiver ahead of transmission, it is clear that no information has been transmitted. Now if each of these bits was independent of each other's value and equally likely to be either $0$ or $1$, we can say 1000 bits( or Shannon's) have been transmitted. 

So we can say that, $\mathbb{X}$ is the set of all possible messages ${x_1, x_2,...,x_n}$ that $X$ could be, and $p(x)$ is the probability of some $x \in X$ , then the entropy $H(X)$ is given by

\begin{equation}
H(X) = \mathbb{E}_X |I(X)| = -\sum_{x \in \mathbb{X}} p(x) \log p(x)
\end{equation}

where

$ I(x)$ = self-information ,i.e., which is the entropy contribution of an individual message

$\mathbb{E}_X$ = Expected value

## The physics of information

### Shannon's Law

Information theory describes in the language of mathematics the limits for the communication of information. These limits are operational, of real engineering significance, and independent of the semantic aspects associated with communication process. They arise from the following set of constraints expressing our inability to appreciate the infinitum:

- Finite energy : Communication occurs by finite expenditure of energy.
- Finite dimensionality : Communication occurs by selecting among a range of possible choices, each identified by a finite number of attributes.
- Finite resolution : Each attribute can be observed with limited precision.

According to Shannon, reliable communication can occur if probability of miscommunication tends to zero within appropriate limiting regimes. In these regimes, some interesting physical phenomena take place such as the following:

*  The space-time fields used to convey information become amenable to a discrete representation, and the finite dimensionality of the physical world is revealed. This allows us to view information-theoretic results as being imposed by laws of nature.
*  Classically information theory considers time asymptotics, used to describe point-to-point communication.
* Their natural counterparts, spatial asymptotics are used to extend the description to communication between multiple transmitters and receivers, and to the remote sensing world around us.

A key measure in information theory is entropy. Entropy quantifies the amount of uncertainty involved in the value of a random variable or the outcome of a random process. For example, identifying the outcome of a fair coin flip (with two equally likely outcomes) provides less information (lower entropy) than specifying the outcome from a roll of a die (with six equally likely outcomes). Some other important measures in information theory are mutual information, channel capacity, error exponents, and relative entropy. Important sub-fields of information theory include source coding, algorithmic complexity theory, algorithmic information theory, and information-theoretic security.

### Concentration Behaviors

Basis of the asymptotic arguments that lead to the information-theoretic limits is the consept of *concentration*.

Let us consider a space-time waveform $f(x,y,z,t)$ of finite energy, transmitted for $T$ seconds. As $T \to \infty$, we can define the effective frequency bandwidth of the waveform as the effective spectral support in the Fourier-transformed angular frequency domain (for more information check out this link to [spectral concentration in wave theory](https://en.wikipedia.org/wiki/Spectral_concentration_problem ) ). So as the time domain support is increased, the signal when viewed in frequency domain can be more and more concentrated inside the bandwidth. Due to this phenomenon EM signals with large $T$ can be considered as occupying a finite bandwidth. 

Signals with finite energy and bandwidth have another mathematical property, i.e., they exhibit a limit on the amount of variation they can undergo in any given time interval, and therefore when observed at a finite resolution, on the amount of information they can carry over time. 

The same limitations apply in the ***spatial domain***. As the region of observation of the signal is stretched by scaling in all of it's coordinates, spectral concentration takes place and this allows the definition of the effective bandwidth in the ***wavenumber domain***, which is the Fourier transform of the spatial domain. This imposes a limit on the number of spatial configurations of the waveform, and thus, when viewed at a finite resolution the amount of information it can carry over space. 

As we are considering space & time asymptotics, another kind of concentration that occurs is, the precision level at which the signal can be observed *probabailistically concentrates* around it's typical value. This allows us to view the uncertainity of the observation as concentrated around it's typical value and determine a resolution limit at which the signal can be observed. 

Combined with the constraints on the form of the signal mentioned above (space & time), it poses an ultimate threshold on the amount of information that can be transported by waves in time and space.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from os import system
```

```python
fig, ax = plt.subplots(nrows = 2, figsize=(12,8))
SAMPLE_RATE = 80 # 1/T. Theoretically, a sample rate of 8 would suffice but due to matplotlib's limitation of
            # plotting curves using a line joining points visualization of sine wave is not possible for 8 samples.   
OBS = 4 # Total Duration of observation 
freq = 4 # Frequency of the generated wave

time = np.linspace(-OBS/2, OBS/2, SAMPLE_RATE * OBS, endpoint=False)

val = np.sin(2 * np.pi * time * freq)

ax[0].plot(time, val, color="red")
ax[0].set_ylabel("Amplitude")
ax[0].set_xlabel("Time")
ax[0].set_title("In time domain")

# N the number of samples taken 

N = SAMPLE_RATE * OBS
fourier_val = fft(val)
fourier_freq = fftfreq(N, 1/SAMPLE_RATE)

if(N%2 == 0):
    ptr = int(N/2 -1)
else:
    ptr = int((N-1)/2)
print(ptr)

# The below split in plotting is done because scipy.fftfreq & scipy.fft returns the values of frequency and 
# fourier transform as 0,1,..., N/2-1, -(N/2-1), ..., -1 (for even N) or 0,1,..., (N-1)/2, -(N-1)/2,..., -1 
# (for odd valued N) 

ax[1].plot(fourier_freq[ptr+1:], fourier_val[ptr+1:], color="blue")
ax[1].plot([fourier_freq[N-1], fourier_freq[0]],[fourier_val[N-1], fourier_val[0]], color="blue")
ax[1].plot(fourier_freq[0:ptr], fourier_val[0:ptr], color="blue")
ax[1].set_ylabel("Power")
ax[1].set_xlabel("Angular Frequency")
ax[1].set_title("In Angular-frequency fourier domain")

plt.subplots_adjust(left=0.1, 
                    bottom=0.1,  
                    right=0.9,  
                    top=0.9,  
                    wspace=0.4,  
                    hspace=0.4) 
```

## The Dimensionality of the Space

---
To quantify the amount of information carried by EM waves of finite energy, we represent the possible message that the transmitter may send using an ensemble of waveforms, and then attempt to quantify the amount of information transferred by correct selection at the receiver of one element from this ensemble, as per Shannon's theory. This gives birth to a fundamental question :

**How many distinct waveforms can we possibly select from?**

The answer depends on the signals' space, or the *number of degrees of freedom* available for communication and on the resolution at which each degree of freedom can be observed. The number of degrees of freedom are limited by the nature of the wave propagation process, and the resolution is limited by the observation process.

### Bandlimitation filtering

Any radiated waveform has an essentially finite bandwidth. Due to interaction with the propagation medium and the measurement appartus (used for signal detection) the higher-frequency get cut off and the signal appears as the output of a linear filter. This filter also shapes the frequency profile, distorting the transmitted waveform. 
An example of this phenomenon is under multiple scattering environment. A transmitted waveform is scattered multiple times creating different copies of the original waveform with different delays and attenuations. The sopies while traveling on different paths might overlap at the receiver creating interference and distorting the original signal. 

In the spatial domain, an analogous effect takes place. The signal observed along any spatial cut-set that separates transmitters and receivers is the image of all the elements radiating from one side of the cut to the other. These waveforms may interfer at the cut-set leading to *spatial filtering*.

### The Number of Degrees of Freedom

The basic property of an EM field allows us to define the size of the signal's space in terms of the number of degrees of freedom. 

Let us consider a one-dimensional, real, scalar waveform $f$ of a single scalar variable $t$. We can assume that $f$ is square-integrable, and

\begin{equation}
\int_{-\infty}^{\infty} f^2(t)dt \leq E
\end{equation}

This ensures that the waveform can be expanded in a series of, possibly complex, orthonormal basis functions $\{\psi_n\}$,
\begin{equation}
f(t) = \sum_{n=1}^{\infty} a_n\psi_n(t),
\end{equation}

where
\begin{equation}
a_n = \int_{-\infty}^{\infty} f(t)\psi_n^*(t)dt
\end{equation}
The equation (7) is  intended in an "energy sens" , i.e., 
\begin{equation}
\lim_{N \to \infty} \int_{-\infty}^{\infty} [f(t) - f_N(t)]^2dt = 0
\end{equation}

where
\begin{equation}
f_N(t) = \sum_{n=1}^{N} a_n\psi_n(t)
\end{equation}

A mathematical explanation for the above equations is, that $f$ lies in ${L}^2(-\infty, \infty)$ and it can be viewed as an infinite dimensional space of coordinates given by the coefficients in eq. (8). By varying the values of these coefficients we can create distinct waveforms and use them to communicate information. If the set of orthonormal basis function $\{\psi_n\}$ is complete, then using eq. (7) we can create any element in the space of signals defined by eq. (6).

We can reasonably expect that only a certain number of coefficients are required to accurately specify any waveform, and going beyond this value doesn't increase the accuracy of the waveforms' representation significantly. It turns out that the smallest value $N$ beyond which varying high-order coefficients doesn't change the waveform significantly has a remarkably precise answer.

Let's consider an observation interval $[-\frac{T}{2}, \frac{T}{2}]$ and introducing the norm to eq. (6), we get, 

\begin{equation}
||f|| = \bigg( \int_{-\infty}^{\infty} f^2(t)dt \bigg)^{1/2}
\end{equation}

The problem translates to determining a value of $N$ such that the approximation error for any signal $f$,
\begin{equation}
||e_N|| = ||f(t) - f_N(t)||
\end{equation}

transitions from being close to maximum to being close to zero.

For signals of spectral support in an interval of size $2\Omega$, and observed at a time interval of size $T$, the angular bandwidth $\Omega$ and the size of observation interval play a key role in determining the number of degrees of freedom as the approximation error undergoes a *phase transition* at the scale of the product $\Omega T$.

So, for any $\epsilon > 0$ letting $N_\epsilon$ be the minimum number of basis functions for which the normalized energy of the error is,
\begin{equation}
\frac{||e_{N_\epsilon}||^2}{E} \leq \epsilon
\end{equation}

and letting 
\begin{equation}
N_0 = \frac{\Omega T}{\pi}
\end{equation}

we have 
\begin{equation}
\lim_{N_0 \to \infty} \frac{N_\epsilon}{N_0} = 1
\end{equation}

**Question 1**. There is no such thing as a "thin shell of small volume." Consider the $N$-dimensional box $C^N(1)$ of side length 1.  Show that all the volume in high dimensions concentrates on the surface of the box; namely, for any $ 0 < \epsilon <1$  

$$ \lim_{N \to \infty} Vol(C^N(1) - C^N(1-\epsilon)) = 1 $$

**Answer.** As the geometry under consideration is a box, and we know that $ Vol(C^N(x)) = x^N$ for an $N$-dimensional box, the above equation can be written as, 

$$ \lim_{N \to \infty} (1)^N - (1 - \epsilon)^N$$

Now, as $N \to \infty$, $(1 - \epsilon)^N \to 0$ and $(1)^N$ stays $1$, we can say that eq. (17) is equivalent to, 

$$ \lim_{N \to \infty} (1)^N - (1-\epsilon)^N = 1 $$

hence we say that eq. (16) is true. A more generalized form can be written as, for an $N$-dimensional box $C^N(x)$ of side length $x$, we can say that, 

$$ \lim_{N \to \infty} (x)^N - (x-\epsilon)^N = x^N , \ \forall \ \epsilon \in (0,1) $$

This can be shown practically by the code below.

```python
import numpy as np
import matplotlib.pyplot as plt

# volume list saves the values
# x is the side length of the box in our specific case 1.
# epsilon is any value between (0,1).
#dimensions is the values of N that we are considering.

volume = []
x = 1
epsilon = 0.002
dimensions = np.arange(1, 2000, 1)
for i in dimensions:
        volume.append(pow(x, i) - pow((x - epsilon), i))
plt.plot(dimensions, volume)
```

**Question 2.** Consider a box shaped shell that gets thinner as $N \to \infty$. Compute for all $a > 0$,

$$ \lim_{N \to \infty} \ Vol(C^N(1) - C^N(1-\frac{a}{N})) $$

**Answer**. Again the volume of a $N$-dimensional box of side length $x$, is given by, $x^N$, hence the equation in the question can be written as, 

$$ \lim_{N \to \infty} \ (1)^N - (1-\frac{a}{N})^N $$

Now, as $N \to \infty$, $\frac{a}{N} \to 0$ and $ (1-\frac{a}{N}) \to 1$, but $ (1-\frac{a}{N})^N \to 0$ hence we can say that, 

$$ \lim_{N \to \infty} \ (1)^N - (1-\frac{a}{N})^N = 1$$, which translates to,

$$ \lim_{N \to \infty} \ Vol(C^N(1) - C^N(1-\frac{a}{N})) = 1 $$

This can be shown using a snippet of code below. 

```python
# The list containing volumes
volume = []
#side length of the box
x = 1
# Value of a, strictly greater than 0
a = 3
# Values of N
N = np.arange(1, 2000, 1)

for i in N:
    volume.append(pow(x, i) - pow((x- a/i), i))
plt.plot(N, volume, color="red")
```
